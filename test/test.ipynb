```python
import networkx as nx

from hiveplot import HivePlot

%matplotlib inline
```

```python
G = nx.read_gpickle('test_graph.pkl')
```

```python
groups = [1,2,3]

nodes = dict()

for g in groups:
    nodes[g] = [n for n, d in G.nodes(data=True) if d['group'] == g]
    
cmap = dict(zip(groups, ['red', 'green', 'blue']))
```

```python
h = HivePlot(nodes, G.edges(data=True), cmap)
h.set_minor_angle()
h.draw()
```
